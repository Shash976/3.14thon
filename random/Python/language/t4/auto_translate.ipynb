```python
import pathlib
import googletrans
import os
```

```python
translator = googletrans.Translator()
locale_dir = pathlib.Path('locales')
full_path = r"C:\Users\shash\Desktop\Code\VSC\test\Python\language\t4\locales"
languages = [l for l in os.listdir("locales") if os.path.isdir(os.path.join(full_path, l))]
```

```python
placeholder_text = r"""
# SOME DESCRIPTIVE TITLE.
# Copyright (C) YEAR THE PACKAGE'S COPYRIGHT HOLDER
# This file is distributed under the same license as the PACKAGE package.
# FIRST AUTHOR <EMAIL@ADDRESS>, YEAR.
#
#, fuzzy
msgid ""
msgstr ""
"Project-Id-Version: PACKAGE VERSION\n"
"Report-Msgid-Bugs-To: \n"
"POT-Creation-Date: 2023-11-27 15:17+0530\n"
"PO-Revision-Date: YEAR-MO-DA HO:MI+ZONE\n"
"Last-Translator: FULL NAME <EMAIL@ADDRESS>\n"
"Language-Team: LANGUAGE <LL@li.org>\n"
"Language: \n"
"MIME-Version: 1.0\n"
"Content-Type: text/plain; charset=UTF-8\n"
"Content-Transfer-Encoding: 8bit\n"

#: main.py:13
msgid "PyQt5 GUI with Labels and Buttons"
msgstr ""

#: main.py:23
#, python-brace-format
msgid "Label {number} with a long text that doesn't fit in one line"
msgstr ""

#: main.py:24
#, python-brace-format
msgid "Button {number} with a long text"
msgstr ""

#: main.py:37
#, python-brace-format
msgid "Button {button_number} clicked!"
msgstr ""
"""
```

```python
import re
def get_final_text(text, language:str="en"):
    text = text[:475] + language + text[475:]
    matches = re.finditer('msgid "(.*)"', text)
    additional = 0
    for match in matches:
        target = text[match.start()+7+additional:match.end()-1+additional]
        if len(target) > 0:
            
            var = re.findall("{(.*)}", target)
            target = re.sub("{.*}", r"  ====  ",target)    
            translation = translator.translate(target, dest=language).text
            var = var if len(var) < 1 else " {" + var[0] + "} "
            target = re.sub(r"====", var, translation ) if type(var) is str else translation
            msgstr = re.search('msgstr "(.*)"', text[match.end():])

            text = text[:msgstr.start()+match.end()+8] + target + text[msgstr.end()-1+match.end():]
            additional += len(target)
    return text
    
```

```python
import io
for language in languages:
    base = os.path.join(full_path, language, "LC_MESSAGES", "base.po")
    text_to_write = get_final_text(text=placeholder_text, language=language)
    print("Writing for lang", language)
    with io.open(base, "w+", encoding='utf-8') as file:
        try:
            file.write(text_to_write)
        except Exception as err:
            print("this is the error, ",err)
            print(text_to_write[650:700])
            continue
    
```
