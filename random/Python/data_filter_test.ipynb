```python
for n, campus in data_frame["Campus"].items():
    if campus=="Pilani":
        x = ''
        for m in data_frame:
            x+= f"{m} - {data_frame[m].iloc[n]}; "
        print(x)
```

```python
def filter(parameters:list, matches:list):
    df = data_frame
    for param_number, parameter in enumerate(parameters):
        match = matches[param_number]
        match = [match] if type(match) is not list else match
        df = df.query(f'{parameter} in {match}')
    print(df)
                

```
