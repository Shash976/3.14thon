```python
import numpy as np
import cv2
import time
```

```python
vid = cv2.VideoCapture(r"C:\Users\shash\Desktop\new1\ml_gui\main\test.h264")
frames = []
ret, frame = vid.read()
while ret:
    ret,frame = vid.read()
    if not ret:
        break
    frames.append(frame.astype(np.uint8))
vid.release()
frames = np.array(frames)
```

```python
def chunks(chunk_size:int):
    start=time.time()
    maxes = []
    number_of_frames = frames.shape[0]
    for  i in range(0, (number_of_frames//chunk_size)*chunk_size, chunk_size):
        chunk = frames[i:i+chunk_size] if number_of_frames > i+chunk_size else number_of_frames-1
        chunk_max = np.max(chunk, axis=0)
        maxes.append(chunk_max)
    maxes= np.array(maxes)
    overall_max = np.max(maxes, axis=0)
    mean = np.mean(overall_max)
    period = time.time() - start
    print("Overall maximum intensity (chunks):", mean)
    print("Time Taken: ", period)
    return period
```

```python
def oneGo():
    start = time.time()
    overall_max = np.max(frames, axis=0)
    mean = np.mean(overall_max)
    period = time.time() - start
    print("Overall maximum value (single):", mean)
    print("time taken ", period)
    return period
```

```python
maximum = 0
def frame_by_frame(frame):
    mean = np.mean(frame)
    global maximum
    if mean > maximum:
        maximum = mean

vectorized_frame_by_frame = np.vectorize(frame_by_frame)

def vectorized():
    start = time.time()
    vectorized_frame_by_frame(frames)
    period = time.time() - start
    print("Overall maximum value (vectorized):", maximum)
    print("time taken ", period)
    return period
```

```python
x = oneGo()
y = chunks(200)
z = vectorized()


print(x>y)
print(y>z)
print(x>z)
```
